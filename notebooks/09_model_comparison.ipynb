```python
# ============================================
# 09 â€” FINAL MODEL COMPARISON
# ============================================

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("seaborn-v0_8")

# --------------------------------------------
# 1. METRICS TABLE
# --------------------------------------------

results = pd.DataFrame({
    "Model": [
        "SARIMA",
        "Prophet",
        "CatBoost",
        "LightGBM"
    ],
    "Type": [
        "Time Series",
        "Time Series",
        "Machine Learning",
        "Machine Learning"
    ],
    "Aggregated_MAPE": [
        8.19,
        4.69,
        2.76,
        2.70
    ],
    "Aggregated_RMSE": [
        335.24,
        223.15,
        120.75,
        118.75
    ]
})

results = results.sort_values("Aggregated_MAPE")

results
```

The empirical comparison reveals a clear performance hierarchy. Classical SARIMA exhibits the weakest forecasting accuracy, while Prophet significantly improves performance by modeling trend and seasonal components. However, machine learning models (LightGBM and CatBoost) substantially outperform both time series approaches. The inclusion of exogenous demand drivers such as price, promotion intensity, and stock availability reduces aggregated forecasting error by approximately 67% relative to SARIMA and 42% relative to Prophet. The close performance between LightGBM and CatBoost confirms robustness of the ML approach.
